### 1. Змінні та типи даних ###

У Python ми працюємо з різними типами даних. Ось основні з них:

int: цілі числа (наприклад, 5, -3, 100).

float: числа з плаваючою комою (наприклад, 3.14, -0.001).

str: рядки (наприклад, "Hello", 'Python').

bool: логічні значення (True або False).

list: списки (наприклад, [1, 2, 3], ["apple", "banana"]).

tuple: кортежі (наприклад, (1, 2, 3)).

dict: словники (наприклад, {"name": "Alice", "age": 25}).


```python
# змінні та типи
x = 5          # int
y = 3.14       # float
name = "John"  # str
is_active = True # bool
```

### 2. Операції ###

Python підтримує стандартні математичні операції:

+ додавання

- віднімання
'
* множення

/ ділення (результат буде типом float)

// цілочисельне ділення (відкидає дробову частину)

% залишок від ділення

** піднесення до степеня

```python
a = 10
b = 3
sum_ab = a + b  # 13
div_ab = a / b  # 3.333333...
floor_div = a // b  # 3
mod_ab = a % b  # 1
power_ab = a ** b  # 1000
```

### 3. Умовні оператори ###

У Python ми використовуємо умовні оператори для виконання різних дій в залежності від умови:

if: для перевірки умови.

elif: для додаткових умов.

else: для дій, які виконуються, якщо жодна умова не була істинною.


```python
age = 18
if age >= 18:
    print("You are an adult")
elif age >= 13:
    print("You are a teenager")
else:
    print("You are a child")
```

### 4. Цикли ###

Цикли дозволяють виконувати певний блок коду кілька разів.

for: цикл для ітерацій по елементах (списки, рядки, діапазони тощо).

while: цикл, що виконується, поки умова істинна.

```python
for i in range(5):  # range(5) дає 0, 1, 2, 3, 4
    print(i)


i = 0
while i < 5:
    print(i)
    i += 1  # збільшення i на 1 в кожній ітерації
```

### 5. Функції ###

Функції дозволяють організувати код у зручні блоки, які можна викликати кілька разів.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))
```

### 6. Списки та їх методи ###

Списки в Python — це змінні, упорядковані колекції елементів. Вони підтримують додавання, видалення елементів, індексацію, зрізи тощо.


```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")  # додаємо елемент
fruits.remove("banana")  # видаляємо елемент
print(fruits[0])  # доступ до елементу за індексом
print(fruits[1:])  # зріз з другого елементу до кінця
```

### 7. Модулі та імпорт ###

Python дозволяє використовувати зовнішні модулі, що допомагають розширити функціональність вашої програми.

```python
import math

print(math.sqrt(16))  # виведе 4.0
print(math.pi)  # виведе значення числа π
```

### 8. Обробка помилок (Exception Handling) ###

У Python можна обробляти помилки за допомогою конструкцій try...except.

```python
try:
    x = 10 / 0  # ділення на нуль викличе помилку
except ZeroDivisionError:
    print("You cannot divide by zero!")
```
