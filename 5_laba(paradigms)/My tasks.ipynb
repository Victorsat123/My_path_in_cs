### Додайте генератор випадкових чисел та викличіть методи deposit та withdraw у циклі передаючи випадкові числа. Виведіть кінцевий результат ###

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  # публічний атрибут
        self.__balance = balance  # приватний атрибут

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance

# Створення рахунку
account = BankAccount("Bohdan", 1000)
print(f"Початковий баланс: {account.get_balance()}\n")
print("-" * 40)

# Виконуємо цикл 5 разів
for i in range(1, 6):
    print(f"--- Ітерація {i} ---")
    
    # Генеруємо випадкову суму для поповнення (наприклад, від 50 до 500)
    random_deposit = random.randint(50, 1000)
    account.deposit(random_deposit)
    print(f"Поповнення (deposit): +{random_deposit}. Баланс: {account.get_balance()}")
    
    # Генеруємо випадкову суму для зняття (наприклад, від 50 до 800)
    random_withdraw = random.randint(50, 600)
    withdraw_result = account.withdraw(random_withdraw)
    
    # Перевіряємо, чи успішно знято кошти
    if withdraw_result == "Insufficient funds":
        print(f"Зняття (withdraw): Спроба зняти {random_withdraw}. Помилка: Недостатньо коштів.")
    else:
        print(f"Зняття (withdraw): -{random_withdraw}. Баланс: {account.get_balance()}")
        
    print() # Порожній рядок для розділення ітерацій

print("-" * 40)
# Виведення кінцевого результату
print(f"Кінцевий баланс: {account.get_balance()}")  
```

### Створіть ще один метод у класі Vehicle та викличіть його з обєкта класу Car ###

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"

    # Додаємо новий метод у базовий клас
    def wash(self):
        return f"Транспортний засіб {self.brand} {self.model} тепер чистий!"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  # виклик конструктора базового класу
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

# Створюємо об'єкт класу Car
car = Car("Toyota", "BMW", 5)
# 1. Виклик методу display_info (перевизначений у Car)
print(car.display_info())

# 2. Виклик НОВОГО методу wash (успадкований від Vehicle)
print(car.wash())
```

### Створіть клас Fish у якого небуде метода speak. Як буде поводити себе обєкт класу Fish при виклику методу speak? ###

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Fish(Animal):
    pass

# Використання:
animals = [Dog(), Cat(), Fish()]
for animal in animals:
    print(animal.speak())
```

### Додайте третій тип зброї — Bow (лук), який:

1) Має власний параметр range_power (дальність);
2) метод attack() з формулою шкоди attack_power + randint(5, 15) + range_power;
3) має метод reload(), який збільшує дальність (range_power += 1); Додайте випадковий вибір зброї з трьох можливих (Sword, Axe, Bow); Зробіть гру покроковою, де користувач може робити хід та вибирати дії з своїм типом зброї (наприклад, атакувати або накласти підсилення щоб підвищити ефективність зброї на наступному кроці); ###

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name:str, health=200): 
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self, another_item):
        pass

    @abstractmethod
    def power_up(self):
        """Абстрактний метод для підсилення зброї (поліморфізм)"""
        pass


class Sword(Item):
    def __init__(self, name, attack_power:int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0
    
    def attack(self, another_item:Item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f" Завдаємо удару мечем {self.name} ({current_attack} шкоди). У супротивника залишилось: {another_item.health} ХП."
    
    def power_up(self): # Замінив sharpening на універсальний power_up
        self._sharp += 10
        return f" Меч {self.name} нагострено! Додаткова шкода: +{self._sharp}"


class Axe(Item):
    def __init__(self, name, attack_power:int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0 # Тепер сокира також використовує цей параметр
    
    def attack(self, another_item:Item):
        # Додав _sharp до формули атаки сокири
        current_attack = self.__attack_power + self._sharp + randint(0, 20)
        another_item.health -= current_attack
        return f" Завдаємо удару сокирою {self.name} ({current_attack} шкоди). У супротивника залишилось: {another_item.health} ХП."

    def power_up(self):
        self._sharp += 10
        return f" Сокира {self.name} розлючена! Додаткова шкода: +{self._sharp}"


class Bow(Item):
    def __init__(self, name, attack_power:int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._range_power = 0 # Власний параметр лука
        
    def attack(self, another_item:Item):
        current_attack = self.__attack_power + self._range_power + randint(5, 15)
        another_item.health -= current_attack
        return f" Стріляємо з лука {self.name} ({current_attack} шкоди). У супротивника залишилось: {another_item.health} ХП."
        
    def reload(self):
        self._range_power += 10 
        return f" Лук {self.name} натягнуто сильніше (reload)! Дальність/шкода: +{self._range_power}"

    def power_up(self):
        # Перенаправляємо універсальний виклик на специфічний метод лука
        return self.reload()


# --- НАЛАШТУВАННЯ ГРИ ---

# Створюємо списки можливої зброї для гравця та ворога
def get_random_weapon(prefix):
    weapons = [
        Sword(f"Ескалібур ({prefix})", 30),
        Axe(f"Кратос ({prefix})", 25),
        Bow(f"Леголас ({prefix})", 20)
    ]
    return choice(weapons)

print("=== ПОЧАТОК ГРИ ===")
player = get_random_weapon("Гравець")
enemy = get_random_weapon("Ворог")

print(f"Ваша зброя: {player.name}")
print(f"Зброя супротивника: {enemy.name}\n")

turn = 1

# --- ІГРОВИЙ ЦИКЛ ---
while player.health > 0 and enemy.health > 0:
    print(f"--- Хід {turn} ---")
    print(f"Ваше ХП: {player.health} | ХП ворога: {enemy.health}")
    print("Ваші дії: 1 - Атакувати | 2 - Підсилити зброю")
    
    # Хід Гравця
    action = input("Виберіть дію (введіть 1 або 2): ")
    
    if action == '1':
        print(player.attack(enemy))
    elif action == '2':
        print(player.power_up())
    else:
        print("Неправильний вибір, ви пропускаєте хід через розгубленість!")
        
    # Перевірка чи ворог переможений
    if enemy.health <= 0:
        print("\n ВИ ПЕРЕМОГЛИ!")
        break
        
    print("-" * 20)
    
    # Хід Ворога 
    enemy_action = choice(['attack', 'attack', 'power_up'])
    if enemy_action == 'attack':
        print(f"[Ворог атакує] {enemy.attack(player)}")
    else:
        print(f"[Ворог підсилюється] {enemy.power_up()}")
        
    # Перевірка чи гравець переможений
    if player.health <= 0:
        print("\n ВИ ПРОГРАЛИ...")
        break
        
    turn += 1
    print("\n" + "="*30 + "\n")
```
