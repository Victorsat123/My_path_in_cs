## Завдання 16

Створити клас AverageCalculator, який:
- Приймає в конструкторі список чисел
- Має метод calculate(), який обчислює середнє арифметичне
- Має property average, яка повертає середнє значення (округлене до 2 знаків)

Не потрібно пояснювати код. Використай форматування Markdown для вставки коду. У функції print використовуй українську мову.

```python
class AverageCalculator:
    def __init__(self, numbers):
        self.numbers = numbers

    def calculate(self):
        if not self.numbers:
            return 0
        return sum(self.numbers) / len(self.numbers)

    @property
    def average(self):
        return round(self.calculate(), 2)

    def __str__(self):
        return "Ім'я: Віктор\nГрупа: КН-32"


if __name__ == "__main__":
    datasets = [
        [10, 20, 30, 40, 55],
        [3.5, 7.2, 1.8, 9.4, 5.0],
        [100, 250, 375, 420]
    ]

    for data in datasets:
        calc = AverageCalculator(data)
        print(calc)
        print(f"Вхідні дані: {data}")
        print(f"Середнє арифметичне: {calc.average}")
        print("-" * 30)
```
