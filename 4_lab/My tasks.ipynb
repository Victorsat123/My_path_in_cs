## Застосуйте методи класу до всіх обєктів a1-a3 ##

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила за мною один рік"


# Створюємо об'єкти
a1 = MyAnimals("Рибки", "Неонка", 1)
a2 = MyAnimals("Кіт", "Джонсон", 5, 10)
a3 = MyAnimals("Собака", "Бомбочка", 2)

# Збираємо всі об'єкти в список
animals_list = [a1, a2, a3]

# Застосовуємо методи класу до всіх об'єктів у циклі
for animal in animals_list:
    print(f"{animal.animals} говорить: {animal.speak()}")
    print(f"{animal.animals} живе зі мною {animal.age} років")
    print(animal.add_one_year())
    print(f"{animal.animals} живе зі мною {animal.age} років\n")
```

## Задайте декілька обєктів в циклі та виведіть глобальну змінну ##

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} якій {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

        # Зміни класу
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1


# --- Створюємо об'єкти в циклі ---
animals_data = [
    ("Рибка", "Булька", 1),
    ("Рибка", "Паучок", 1),
    ("Рибка", "Джордж", 1),
    ("Кіт",   "Шрек", 5, 10),
]

objects = []   # сюди збережемо всі створені об'єкти

for data in animals_data:
    obj = MyAnimals(*data)
    objects.append(obj)

# Виведення глобальних змінних класу
print(f"\nВ мене зараз є тваринки: {list(MyAnimals.total_animal_types)}")
print(f"Їх кількість: {MyAnimals.total_animals}")

# --- Демонстрація різниці між змінною класу та змінною об'єкта ---
a1 = objects[0]

print(f"\nЗначення змінної класу через об'єкт a1: {a1.total_animal_types}")

# Перезаписуємо змінну класу локально в об'єкті
a1.total_animal_types = {"Собака"}

print(f"Після зміни в об'єкті:")
print(f"  a1.total_animal_types -> {a1.total_animal_types}  (локальна змінна об'єкта)")
print(f"  MyAnimals.total_animal_types -> {MyAnimals.total_animal_types}  (справжня класова змінна)")
```

## Викличте статичний метод для кожного обєкта a1-a2 ##


```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила за мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

# Виклик статичного методу для кожного об'єкта
print(f"Статичний метод для a1: {a1.call_pet()}")
print(f"Статичний метод для a2: {a2.call_pet()}")

# Також можна викликати через клас
print(f"Статичний метод через клас: {MyAnimals.call_pet()}")
```

## Додайте ще одну properties та виведіть її ##

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    # Вік до якого тварина буде жити
    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

    # Нова property: скільки років залишилось
    @property
    def years_left(self):
        return self.max_age - self.age

    def add_one_year(self):
        self.age += 1
        return f"Моя тваринка {self.animals} прожила за мною один рік"


# Створюємо об'єкт
a1 = MyAnimals("Кіт", "Боксер", 5)

# Вивід обох property
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Мій {a1.animals} має залишилось {a1.years_left} років")

# Додаємо рік
a1.add_one_year()

# Перевіряємо знову
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Мій {a1.animals} має залишилось {a1.years_left} років")
```

## Допишіть метод класу який буде превіряти що атрибут name заданий з великої букви ##

```python
class MyAnimals:
    total_csv = 0
    total_objects = 0

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    # Метод класу для перевірки, що name з великої літери
    @classmethod
    def check_name_capital(cls, obj):
        if obj.name and obj.name[0].isupper():
            return True
        return False


# --- Створюємо об'єкти ---
a1 = MyAnimals.from_csv("Кіт,Шрек,5")
a2 = MyAnimals("Собака", "бобер", 2)  # маленька буква для прикладу

# --- Вивід ---
print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість об’єктів: {MyAnimals.total_objects}, створених з CSV: {MyAnimals.total_csv}")

# --- Перевірка імен ---
print(f"Ім’я {a1.name} з великої букви? {MyAnimals.check_name_capital(a1)}")
print(f"Ім’я {a2.name} з великої букви? {MyAnimals.check_name_capital(a2)}")
```
